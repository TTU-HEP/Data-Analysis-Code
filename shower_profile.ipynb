```python
import sys
import numpy as np
import matplotlib.pyplot as plt
import csv
import ROOT

# ROOT.gStyle.SetOptStat("emr")
ROOT.gStyle.SetOptStat("")

inFileName="mc_Dream_run04_00_Test_2000ev_pi+_50.0_50.0.root"
infile=ROOT.TFile(inFileName,"READ")

yscalefactor=1.0/2000.0

h1={}

nbins=100
xmin=0.0
xmax=200.0

s="zprofile"
h1[s]=ROOT.TH1D(s,s,nbins,xmin,xmax)

h1["Cz"]=infile.Get("ncerCz")
nx=h1["Cz"].GetNbinsX()
for ix in range(1,nx+1,1):
    xval=(h1["Cz"].GetBinCenter(ix))*2.0
    yval=(h1["Cz"].GetBinContent(ix))*yscalefactor
    print("ix",ix,xval,yval)
    h1["zprofile"].Fill(xval,yval)

nbins=100
xmin=0.0
xmax=5.0

s="timeprofile"
h1[s]=ROOT.TH1D(s,s,nbins,xmin,xmax)

h1["Ct"]=infile.Get("ncerCt")
nx=h1["Ct"].GetNbinsX()
for ix in range(1,nx+1,1):
    xval=(h1["Ct"].GetBinCenter(ix))*0.05-6.2
    yval=(h1["Ct"].GetBinContent(ix))*yscalefactor
    # print("ix",ix,xval,yval)
    h1["timeprofile"].Fill(xval,yval)
```

```python
c2 = ROOT.TCanvas("myCanvasName","The Canvas Title",800,600)
c2.Divide(1,1)
c2.cd(1)
h1["zprofile"].SetLineColor(1)
h1["zprofile"].SetLineWidth(2)
h1["zprofile"].SetTitle("shower profile (z)")
h1["zprofile"].GetXaxis().SetTitle("z, cm")                         
h1["zprofile"].GetYaxis().SetTitle("Cherenkov Photons") 
h1["zprofile"].GetXaxis().SetTitleSize(0.05)
h1["zprofile"].GetYaxis().SetTitleSize(0.05)
h1["zprofile"].GetXaxis().SetTitleOffset(0.9)
h1["zprofile"].GetYaxis().SetTitleOffset(0.9)
h1["zprofile"].Draw("HIST")

c2.Draw()
pdfname="shower_profile_z.pdf"
c2.SaveAs(pdfname)
```

```python
c3 = ROOT.TCanvas("myCanvasName","The Canvas Title",800,600)
c3.Divide(1,1)
c3.cd(1)

h1["timeprofile"].SetLineColor(1)
h1["timeprofile"].SetLineWidth(2)
h1["timeprofile"].SetTitle("shower profile (time)")
h1["timeprofile"].GetXaxis().SetTitle("time, ns")                         
h1["timeprofile"].GetYaxis().SetTitle("Cherenkov Photons")
h1["timeprofile"].GetXaxis().SetTitleSize(0.05)
h1["timeprofile"].GetYaxis().SetTitleSize(0.05)
h1["timeprofile"].GetXaxis().SetTitleOffset(0.9)
h1["timeprofile"].GetYaxis().SetTitleOffset(0.9)
h1["timeprofile"].Draw("HIST")

c3.Draw()
pdfname="shower_profile_time.pdf"
c3.SaveAs(pdfname)
```
